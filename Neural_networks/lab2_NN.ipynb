Для решения задачи создадим класс перцептронная с возможностью мультиклассовой классификации

```python
import numpy as np

class MultClassPerceptron:
  def __init__(self, ep = 100, lr_rate = 0.1, num_classes = 2, k_true = 1, k_false = 1):
    self.ep = ep
    self.lr_rate = lr_rate
    self.num_classes = num_classes

    # k_true указывает на скорость обучения для нейрона который выдал верный ответ
    # k_false указывает на скорость обучения для нейрона который выдал неверный ответ
    self.k_true = k_true
    self.k_false = k_false

  def fit(self, X, y):
    num_features = X.shape[1]

    # Сформируем массив весов, где каждый столбец будет отвечать за веса соответствующего нейрона
    self.w = np.ones([num_features, self.num_classes])
    for i in range(self.ep):
      for sample, answer in zip(X, y):
        pred = self.predict_sample(sample)
        for neuron in range(self.num_classes):
          if pred == answer and pred == neuron:
            w_update = self.k_true * self.lr_rate
          else:
            w_update = self.k_false * self.lr_rate

          self.w[:, neuron] += w_update * sample

  def predict_sample(self, sample):
    pred = np.ones(self.num_classes)
    for neuron in range(self.num_classes):
      sum = np.dot(sample, self.w[:, neuron])
      pred[neuron] = sum
    # функция возвращает номер нейрона с максимальным значение
    # это число номер предсказанного класса
    return pred.argmax()

  def predict(self, X):
    pred = []
    for sample in X:
      pred.append(self.predict_sample(sample))
    return pred

  # Создадим функцию-оценьщика по метрике accuracy
  def score(self, X, y):
      pred = self.predict(X)
      accuracy = np.sum(np.where( pred == y , 1, 0))/ len(y) * 100
      return accuracy
  def w_get(self):
    return self.w

```

```python
#Преабразуем датасет к np.array
dataset = np.array(dataset)

#Сформируем обучающую выборку
X_train = dataset[:, 0:3]
y_train = dataset[:, 3]

#Создадим модель
clf = MultClassPerceptron(ep=10, lr_rate=0.5, num_classes=3, k_true=3, k_false=1)

#Обучим ее на нашей выборке
clf.fit(X_train, y_train)

#Посмотрим на получившиеся веса
clf.w_get()
```

Посмотрим на точность нашей модели
